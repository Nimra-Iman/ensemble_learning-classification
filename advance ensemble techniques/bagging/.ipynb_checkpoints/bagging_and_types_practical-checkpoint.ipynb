```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_moons
x,y=make_moons(n_samples=1000, noise=0.2)
dataset=pd.DataFrame(x, columns=['a','b'])
dataset['target']=y
dataset.head()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='a',y='b', data=dataset, hue='target')
plt.show()
```

```python
x=dataset.drop(columns='target', axis=1)
y=dataset['target']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
```

# bagging meta classifier

```python
from sklearn.svm import SVC
svmm=SVC()
```

```python
bagging_meta_classifier=BaggingClassifier(estimator=svmm,n_estimators=50) #ab n yhan estimator
# m khud s svm use kia, agr na 
# kru to DT automatically None ki jga a jay ag, n estimators equals 50, means 50 independent
# models, in this case, 50 SVM models are trained. 
```

```python
bagging_meta_classifier.fit(x_train, y_train)
```

```python
bagging_meta_classifier.score(x_train, y_train)*100, bagging_meta_classifier.score(x_test, y_test)*100
```

# random forest

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=20)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=45,random_state=42,max_depth=5)
rf.fit(x_train, y_train)
```

```python
rf.score(x_train, y_train)*100,  rf.score(x_test, y_test)*100
```
