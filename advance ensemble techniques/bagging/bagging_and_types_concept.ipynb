# WHY we use random forest instead of decision tree?

```python
if we use decision tree without pruning then we face overfitting which means high training accuracy
and low testing accuracy. 
--> high training accuracy ko hm low bias 
--> low testing accuracy ko hm high variance bolty hn

decision tree s low bias and high variance acheive hota h or hm is high variance ko easily random
forest s low variance m 
convert kr dety hn q k hmara goal is to achieve low variance and low bias, ab random forest m 
hm bagging ko use krty hn,
yani differenet base models train krty h dataset k sub-parts pr yani bht saary decision trees
bnty hn to is trah s high variance low variance m convert ho jata h
```

```python
Q:            is all sub-parts have same size?
 Usually they have the same size but with some parameters we can change the size
of different sub-parts. jo subsets banty hn un m esa hota h k agr original data m 5 
rows hn (1,2,3,4,5), then sample jo bny ga also have 5 rows may be, 1st sample has (1,1,2,4,1), 2nd
sample has (2,3,1,5,5) and so on,,,  THIS IS FOR BOTH META ESTIMATOR AND RANDOM FOREST

    
    In bagging meta estimator there is only row sampling by default,, but we can do feature sampling
as well using parameter max_features: BaggingClassifier(max_features=0.7)

In random forest it also does row sampling as well as feature sampling by default. 
    
```

```python
     decision trees mostly overfit because it does a lot of different branches. A decision tree
tries to make the tree in such a way that at each leaf node there is only one sample present.
    It means that it does a lot of overfitting. When we make a random forest, different 
decision trees are made. These decision trees are also overfitted but but random forest reduced 
the variance by averaging predictions at the end. 
```

```python
Bootstrap samples are usually the same size as the original dataset.
Subsets do not have to be smaller; duplicates are allowed.
Bagging normally uses row sampling only.
Random Forest uses both row sampling and feature sampling.
Individual trees in a Random Forest can still overfit.
Random Forest reduces overfitting by averaging many diverse trees, which reduces variance,
not because each tree learns less information.
```
