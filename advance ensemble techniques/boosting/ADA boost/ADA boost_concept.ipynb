```python
---> saari rows ko is tree pr test kryn gy
ab dekha hm n yhan best tree stump select kia h, y output esy predict kr rha h k jab bhi C==G hua to 
YES vali side pr ao and then is yes vali side ki branck k uper 3YES/0NOs likha h to yani is vali row ki
output YES ay gi, but problem a rhi h 2nd last (highlighted) row mjhan C==N h yani y NO vali side pr 
jay ga or us k uper 1YES/3NOs likha y yani is row ki output NO honi chaye but asal m YES h

--> to hm is step m phly to jitny errors is stump n create kiye un ka sum kryn gy: yhan answer 1/7 
ay ga q k just ek hi wrong 
prediction hua, agr 2 wrong rows wrong predict hoti to hm 2/7 kr dety  

---> phir hm is stumpt ki performance check kryn gy
performance of stump = 1/2 * ln(1-TE/TE) 
which is 0.896 in this case and  TE is total error
```

# step : 4

```python
step:4 update weights for correctly and incorrectly classified points
----> hm vo data points jo k ghlat predict huy un ka weights increase kryn gy and jo shi predict
huy un ka weights
decrease kryn gy ta k next model ko train hony k liye hm n jo data points pick krny hn to un ghlt
predicted data points(jo pichly model n ghalat predict kiye) k pick krny ki probablility 
increase ho jay
```

```python
so: formula to update weights(actually decrease weights) for correctly classified points:
 = weight * (e^(-performance of stump))
 = 1/7    * (e^(-0.896))
 = 0.058
```

```python
so: formula to update weights(actually increase weights) for incorrectly classified points:
 = weight * (e^(performance of stump))
 = 1/7    * (e^(0.896))
 = 0.349
```
