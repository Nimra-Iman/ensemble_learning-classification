### 2--- Optimized data structures

```python
Parallel processing achieved only due to optimized data structres. in other ML models, data 
structures store data in row wise blocks but in XG_Boost, data structures store data in column vise
blocks. And when Parallel processing is required, working on each column vis bloxk start. 
--< and this is only 1 example, there are also many other examples where smartly work with data 
structures
```

### 3--- Cashe awareness

```python
CPU and Ram interact with each other, jo cheezy bhhhtt bar bar CPU m lai ja rhi hoti hn vo CPU m 
cashe menory m store ho jati h, ta k baar baar RAM s request na krni pry. XG_Boost isi cash k 
concept ko use krta h

XG_Boost used histogram based training apporach in tree building, and histogram is based on BINS. 
If our feature's value range from 50 to 150. Different bins created such as one is from 50 to 60 ,
then 60-70 and so on. It stores histogram of that featrure in cache memory because they are 
frequently used in tree bulding. Moreover, it also store other things that will frequently use ,
so speed increase in these ways.
```

### 4--- Out of core computing

```python
problem arises when you have dataset of 10 GB and your system's RAM is only of 8GB. XG_Boost still
work, it divide our dataset into chunks let's say it has divided it into 5 chunks each of 2GB. 
each chunk is loaded in memory sequentially and training takes place. When one chunk is completed, 
training resumes from next chunk
```

### 5--- Distributed Computing

```python
You have a lot of computters , each called node, and you distribute your work in different nodes. 
    Each node performs its work parralley, this is distributed computing. 

---> here, if you have a very big dataset, and you cannot even load your data on single device,
you can divide it into different parts and can train on
different devices, each device train independently and results are aggregated at end.
means each device do its own splitting creiteria and ek master node hota h jo k saari devices 
s un ka apna apna slitting crietrian puchta h and and m ek final criteria decide krta h
and jesy hi decide ho gya to bs hmara kaam ho gya, yani ek final DT a gya jis pr testing 
kr skyn gy
this has more advantages than out of core computing because in out of core computing, this work
occures sequentially and in distributed computing, this occurs parrallely.
```
