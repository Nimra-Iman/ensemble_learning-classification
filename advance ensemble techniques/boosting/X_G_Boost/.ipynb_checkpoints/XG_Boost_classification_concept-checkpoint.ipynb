```python
* BASE MODEL: we build a base model and base model will buld on basis of logg(odds) and formula for it
is log(p/(1-p)) where p is probability of positive class which is 1 in our case.

--> so by considering the following dataset: outpuit of base model will be: log((3/5)/(2/5)) as 3/5 is actually the 
probability of 1, so output of base model will be: 0.405
```

```python
* NOW WE WILL CONSTRUCT DT ON BASIS OF CGPA AND RES1
--> DT is same but phly hm similarity score of root calculate kr lety hn but formula for it is little bit different
from SS for regression, formula is : 
            (sum of residuals)² / (sum of previous probability * (1- sum of previous probability) * LR)
sum of previous probability is actually output of base model in terms of probablity ( which is actually 0.6+0.6+0.6+0.6+0.6)

so, SSroot will be: 0
```
