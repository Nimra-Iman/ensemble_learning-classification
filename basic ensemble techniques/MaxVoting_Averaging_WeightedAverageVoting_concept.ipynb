```python
it is used in both classification and regression, where hm models ko weightage dety hn, yani e.g: hm 
n SVM ko 4 dia and KNN ko 0.5 dia and DT to 2 dia, weght deny s model ka influnece prediction krty
huy itna barh gya jitna us ko weight dia tha, yani jis ka weight ziada us ka influence ziada

In classification if we have three models A, B, and C, the weight of model A is 0.5,
the weight of model B is 0.3, and the weight of model C is 0.1, Model A predicted as spam while
Model B and C predicted as not spam. 
For spam the total weight would be 0.5 and for not spam the total weight will be 0.3 + 0.1 = 0.4. 
    Now the total output would be spam although it is predicted by only one model. The other
two models predicted it as not spam but their total weight for not spam is less than the 
weight of predicting it as spam so therefore spam is the output. 
```
